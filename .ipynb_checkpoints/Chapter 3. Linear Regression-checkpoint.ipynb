### 8

```python
auto = pd.read_csv('data/auto.csv')
auto.head()
```

```python
model = smf.ols("mpg ~ horsepower", data=auto).fit()
model.summary()
```

<b>
i) Yes there is strong evidence of a negative relatonship between mpg and horsepower

ii) the relationship between mpg and horsepower is strong as p-value ~ 0.

iii) negative

iv) for horsepower of 98 the mpg is approximately 24.5. 

</b>

```python
# Plot of the Regression

fig,ax = plt.subplots(figsize=(8,6))
ax.scatter(auto['horsepower'].values,auto['mpg'].values,  edgecolors='b', label="data")
ax.plot(auto['horsepower'].values, model.fittedvalues, 'g', label="OLS")
ax.legend(loc='best');
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,4))

# Residuals
ax1.scatter(model.fittedvalues, model.resid, facecolors='none', edgecolors='b')
ax1.set_xlabel('fitted values')
ax1.set_ylabel('residuals')


# Studentized Residuals
#Studentizing is an anology of standartizing or normalizing.
studentized_residuals = OLSInfluence(model).resid_studentized_internal
ax2.scatter(model.fittedvalues, studentized_residuals, facecolors='none', edgecolors='b')
ax2.set_xlabel('fitted values')
ax2.set_ylabel('studentized residuals')

leverages = OLSInfluence(model).influence
ax3.scatter(leverages, studentized_residuals, facecolors='none', edgecolors='b')
ax3.set_xlabel('Leverage')
ax3.set_ylabel('studentized residual')

plt.show()
```

The residuals show a V shape which indicates a poor fit at the extremes of the fitted data.
Studentized residuals and Leverage plots show that a quadtratic fit is more appropriate.

```python

# Construct model estimate and summarize
model = sm.OLS(auto['mpg'],sm.add_constant(auto.iloc[:,1:-1])).fit()
print(model.summary())
```

<b>
    
i.) There is a clear relationship between predictor and response F-statistic is very high. 

ii.) weight, displacement, year and origin have small p-value.

iii.) The higher the year the more the mpg.
</b>

```python
results_influence = OLSInfluence(model)

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(12,10))

ax1.scatter(model.fittedvalues, model.resid)
ax1.set_ylabel("Raw Residuals")
ax3.scatter(model.fittedvalues, results_influence.resid_studentized_external)
ax3.set_ylabel("Studentized Residual")
sm.graphics.qqplot(model.resid / np.sqrt((model.resid ** 2).sum() / 390), line='45', ax=ax2)
ax4.scatter(results_influence.resid_studentized_external ** 2, results_influence.influence);
```

The qqplot indicates that there are some outliners in the right tail.The graphs also display a similar U-shape ndicating that the fit is worse at the extremes.

```python
from statsmodels.graphics.regressionplots import plot_leverage_resid2
fig, ax = plt.subplots(figsize=(8,6))
fig = plot_leverage_resid2(model, ax = ax)
```

Point 13 has a very big leverage.

### 10

```python
carseats = pd.read_csv('data/carseats.csv')
carseats.head()
```

```python
model = smf.ols('Sales ~ Price + Urban + US', data=carseats).fit()
model.summary()
```

<b><i> Only US and Price predictors are statistically significant. A high p-value for the Urban variable indicates that it does not influence the Sales. For every unit increase in Price there is a 0.0545 decrease in Sales.  </i></b>

 ### The equations:
    
<ol> 
<li> For US:  <i> Sales = 13.0435 + 1.2006 - 0.0545*Price <i> </li>
    <li> Non-US: <i> Sales = 13.0435 - 0.0545*Price </i> </li>
</ol>
    
<b> We reject the null hypothesis for US and Price. </b>

```python
# e) a smaller model without urban
model = smf.ols('Sales ~ Price + US', data=carseats).fit()
model.summary()

```

<b> Both the initial and the smaller models have a very small and identical R-squared. </b>

```python
fig, ax1 = plt.subplots(1,1,figsize=(8,4))

# Residuals
ax1.scatter(model.fittedvalues, model.resid, facecolors='none', edgecolors='b')
ax1.set_xlabel('fitted values')
ax1.set_ylabel('residuals')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
fig = plot_leverage_resid2(model, ax = ax)
```

<b> There are not any significant outliners in the residuals. In the Leverage graph only few point are bigger that 0.025. <b>

```python
np.random.seed(1)
x1 = np.random.rand(100)
x2 = .5 * x1 + np.random.rand(100) / 10
y = 2 + 2 * x1 + .3 * x2 + np.random.randn(100)
```

```python
# the coefficients ar highly correlated
np.corrcoef(x1, x2)
```

```python
plt.scatter(x1,x2)
plt.show()
```

```python
X = np.column_stack((np.ones(100), x1, x2))
model = sm.OLS(y, X).fit()
model.summary()

# Both variables are  not significant
```

```python
X = np.column_stack((np.ones(100), x1))
model = sm.OLS(y, X).fit()
model.summary()
```

```python
X = np.column_stack((np.ones(100), x2))
model = sm.OLS(y, X).fit()
model.summary()
```

<b> Since x1 and x2 are highly correlated the R-squared of the models are also similar. 
In the first model x1 is positive and x2 is negative, which does not seem right. Also, since 
x1 and x2 are highly correlated the model is affected by colinearity, which will cause the standard
error of the predictors to grow, which leads to their insignificance.
</b>

```python
x1_new = np.append(x1, .1)
x2_new = np.append(x2, .8)
y_new = np.append(y, 6)

X = np.column_stack((y_new, x1_new, x2_new))
df = pd.DataFrame(X, columns=['y', 'x1', 'x2'])
model = smf.ols('y ~ x2', data=df).fit()
model.summary()

```

```python
# the correlation decreases
np.corrcoef(x1_new, x2_new)
```

```python
boston = pd.read_csv('data/boston.csv')
boston.head()
```

```python
# a.) A simple linear regression of each predictor

for col in boston.columns[1:]:
    model = smf.ols('crim ~ {}'.format(col), data=boston).fit()
    print(model.conf_int())
```

```python
formula = "crim ~" + " + ".join(boston.columns[1:])
formula
```

```python
# the model with all the variables
model = smf.ols(formula, data=boston).fit()
model.summary()
```

```python
# twe select only the significant predictors
model.tvalues[abs(model.tvalues) > 2]
```

```python
# we select the parameters without the intercept
multiple_regression_params = model.params.iloc[1:]

simple_params = pd.Series()

for col in boston.columns[1:]:
    simple_model = smf.ols('crim ~ {}'.format(col), data=boston).fit()
    simple_params = simple_params.append(simple_model.params.loc[[col]])
    
merged_models = pd.DataFrame({'simple': simple_params, 'multiple':multiple_regression_params})
merged_models
```

```python
merged_models.plot.scatter('simple', 'multiple')
plt.show()
```

```python
# We check the significance of the variables raised to power 2 and 3. 
# the coeff of the nox predictor is way to high.
for col in boston.columns[1:]:
    model = smf.ols('crim ~ {} + np.power({}, 2) + np.power({}, 3)'.format(col, col, col), data=boston).fit()
    print(model.conf_int())
```
